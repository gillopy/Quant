```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Descargar datos de una acción (puedes cambiar el ticker y el rango de fechas)
ticker = 'GGAL.BA'  # Cambia esto por el ticker de tu interés
data = yf.download(ticker, start="2024-01-01", end="2024-10-25")

# Calcular los retornos porcentuales diarios
data['Retorno'] = data['Adj Close'].pct_change() * 100  # Retorno diario en porcentaje

# Filtrar los días que subió más de un 5%
dias_subida_5 = data[data['Retorno'] > 5]

# Filtrar los días que bajó más de un 5%
dias_bajada_5 = data[data['Retorno'] < -5]

# Graficar el precio de cierre con los días resaltados
plt.figure(figsize=(10, 6))

# Gráfico de línea para los precios de cierre
plt.plot(data.index, data['Adj Close'], label='Precio de Cierre', color='blue')

# Marcar los días que subieron más del 5%
plt.scatter(dias_subida_5.index, dias_subida_5['Adj Close'], color='green', label='Subidas > 5%', marker='^', s=100)

# Marcar los días que bajaron más del 5%
plt.scatter(dias_bajada_5.index, dias_bajada_5['Adj Close'], color='red', label='Bajadas > 5%', marker='v', s=100)

# Añadir etiquetas y título
plt.title(f'{ticker}: Días con Subidas y Bajadas mayores al 5%')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
