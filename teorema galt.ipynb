```python
import matplotlib.pyplot as plt
import numpy as np

# Parmetros del teorema
C = 200000  # capital inicial
r = 0.08    # retorno anual de la inversin
m = 0.02    # mantenimiento anual (solo aplica a propiedad comprada)
R = 1000    # alquiler mensual
t_years = np.arange(1, 21)  # horizonte de 1 a 20 aos

# Frmulas del teorema de Galt
ganancia_invertir = C * ((1 + r)**t_years - 1) - C * m * t_years
costo_alquiler = R * 12 * t_years

# Visualizacin
plt.figure(figsize=(10, 6))
plt.plot(t_years, ganancia_invertir, label='Ganancia por invertir', linewidth=2)
plt.plot(t_years, costo_alquiler, label='Costo total de alquiler', linewidth=2, linestyle='--')
plt.axhline(0, color='black', linewidth=0.8, linestyle=':')
plt.xlabel('Aos')
plt.ylabel('USD acumulado')
plt.title('Teorema de Galt: Alquilar e Invertir vs Comprar')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
